```python
from numpy import expand_dims
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.preprocessing.image import ImageDataGenerator
from matplotlib import pyplot
import cv2
import os
from pathlib import Path
```

```python
dataset = Path('Flask_Practice/static/Dataset/')
```

```python
count = 0
image_data = {}
for img_path in dataset.glob("*"):
    image_data[str(img_path).split("\\")[-1]] = cv2.imread(str(img_path).split("/")[-1])
```

```python
class ImageGen:
    
    image_data = None
    def __init__(self,location,no_of_image):
        self.location = location
        self.no_of_image = no_of_image
    
    def load_images(self):
        global image_data
        image_data = [['Unknown']]
#         for img_path in self.location.glob("*"):
#             image_data.append([str(img_path).split("\\")[-1]])
        print(image_data)
        
        
    def Rotate(self):
        file_types = ['*.jpg','*.jpeg']
        for user_img in image_data:
            count = 0
            for file in file_types:
                for img_path in Path(str(self.location)+"\\"+str(user_img[0])+"\\").glob(file):

                    img = load_img(str(img_path))
                    data = img_to_array(img)
                    samples = expand_dims(data, 0)
                    datagen = ImageDataGenerator(rotation_range=90)
                    it = datagen.flow(samples, batch_size=1)

                    
                    file_name =  str(img_path).split("\\")[-1]
                    file_name = file_name.split(".")[0]
                    path = "\\".join(str(img_path).split("\\")[:-1])
                    print(str(file_name)+"Rotation")
                    for i in range(5):
                        batch = it.next()
                        image = batch[0].astype('uint8')
                        
                        cv2.imwrite(path+"\\"+str(file_name)+str(count)+".jpg",cv2.cvtColor(image, cv2.COLOR_RGB2BGR))
                        count+=1
#                         self.no_of_image +=1
            
    
    def Zoom(self):
 
        file_types = ['*.jpg','*.jpeg']
        for user_img in image_data:
            count = 0
            for file in file_types:
                for img_path in Path(str(self.location)+"\\"+str(user_img[0])+"\\").glob(file):
                    
                    img = load_img(str(img_path))
                    data = img_to_array(img)
                    samples = expand_dims(data, 0)
                    datagen = ImageDataGenerator(zoom_range=[0.8,1.0])
                    it = datagen.flow(samples, batch_size=1)
                    
                    path = "\\".join(str(img_path).split("\\")[:-1])
                    file_name =  str(img_path).split("\\")[-1]
                    file_name = file_name.split(".")[0]
                    print(str(file_name)+"Zoom")
                    for i in range(5):
                        batch = it.next()
                        image = batch[0].astype('uint8')
                        cv2.imwrite(path+"\\"+str(file_name)+str(count)+".jpg",cv2.cvtColor(image, cv2.COLOR_RGB2BGR))
                        count+=1
#                         self.no_of_image +=1
#                         if(self.no_of_image >= 200):
#                             return 
                
                
    def Brightness(self):
        
        file_types = ['*.jpg','*.jpeg']
        for user_img in image_data:
            count = 0
            for file in file_types:
                for img_path in Path(str(self.location)+"\\"+str(user_img[0])+"\\").glob(file):
                    
                    img = load_img(str(img_path))
                    data = img_to_array(img)
                    samples = expand_dims(data, 0)
                    datagen = ImageDataGenerator(brightness_range=[0.5,1.0])
                    it = datagen.flow(samples, batch_size=1)
                    
                    file_name =  str(img_path).split("\\")[-1]
                    file_name = file_name.split(".")[0]
                    path = "\\".join(str(img_path).split("\\")[:-1])
                    print(str(file_name)+"Brightness")
                    for i in range(5):
                        batch = it.next()
                        image = batch[0].astype('uint8')
                        cv2.imwrite(path+"\\"+str(file_name)+str(count)+".jpg",cv2.cvtColor(image, cv2.COLOR_RGB2BGR))
                        count+=1
#                         self.no_of_image +=1
#                         if(self.no_of_image >= 200):
#                             return 
        
        
```
